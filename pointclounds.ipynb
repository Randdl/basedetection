```python
import numpy as np
```

```python
path = '../Kitti/raw/training/velodyne/000000.bin'
points = np.fromfile(path)
points = points.reshape((-1, 4))
print(points)
```

```python
calib = {}
with open('../Kitti/raw/training/calib/000000.txt') as inp:
    for line in inp.readlines():
        if ':' in line:
            key, value = line.split(':', 1)
            calib[key] = np.array([float(x) for x in value.split()])
P2_rect = calib['P2'].reshape(3, 4)
```

```python
points_1 = points[:, 0:3]
points_1 = np.transpose(points_1)
points_1 = np.vstack((points_1, np.ones((points_1.shape[-1]))))
points_2D = P2_rect.dot(points_1)
points_2D = points_2D / points_2D[2]
points_2D = np.rint(points_2D).astype(int)
print(points_2D)
```

```python
depth = np.zeros((1224, 370))
for i in range(points_2D.shape[1]):
    x = points_2D[0, i]
    y = points_2D[1, i]
    if -1 < x < 1224 and -1 < y < 370:
        depth[x, y] = np.sqrt(np.square(points[i, 0:3]).sum())
        print(x, y)
        print(np.sqrt(np.square(points[i, 0:3]).sum()))
print(depth)
```
